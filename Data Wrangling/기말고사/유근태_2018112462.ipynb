```python
import seaborn as sns
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

# 1-1)

```python
titanic = sns.load_dataset('titanic')

titanic.info()
```

```python
# 자동화 코드
def drop_missing_col(df):
    max = 0
    for col in df.columns:
        if df[col].isnull().sum() > max:
            max = df[col].isnull().sum()
            col_name = col

    df.drop(col_name, axis=1, inplace=True)

    return df

titanic = drop_missing_col(titanic)
titanic
```

```python
# 누락 데이터 개수 출력
titanic.isnull().sum()
```

```python
url = 'https://www.wizwid.com/display/best'

res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
div = soup.find('div', class_ = 'product_list01')

neme_list = div.select('div.product_info > div.p_name > a')
price_list = div.select('div.product_info > div.price > a > span.price01')

real_name = []
real_price = []

for name in neme_list:
    real_name.append(name.text)

for price in price_list:
    real_price.append(price.text)
```

```python
price = soup.find_all('div', class_ = 'price')

sale_list = []

for d in price:
    if d.find('span', attrs = {'class' : 'price03'}) == None:
        sale_list.append('-')

    else:
        sale_list.append(d.find('span', attrs = {'class' : 'price03'}).text)

len(sale_list)
```

```python
df = pd.DataFrame({'제품명' : real_name, '가격' : real_price, '할인(%)' : sale_list})
df
```
