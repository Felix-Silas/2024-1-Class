```python
# xml.etree.ElementTree 모듈 사용
import xml.etree.ElementTree as ET

# parse xml file
tree = ET.parse('mid-term-rss3.xml') # tree: 전체 XML 객체

# get root node
root = tree.getroot() # root는 xml 문서의 최상단 루트 태그를 가리킴
```

```python
root
```

```python
# root 하위의 Data/Observation의 자식 중에 “Dim“과 일치하는 모든 태그를 리스트로 리턴
list_dim = root.find('channel/item/description/body/location/')

for list in list_dim:
    print(list.attrib)
#list_dim[8].attrib
```

```python
location_list = root.findall('channel/item/description/body/location')

for location in location_list:
    city_list = location.findall('city')
    tmEf_list = location.findall('data/tmEf')
    wf_list = location.findall('data/wf')
    tmn_list = location.findall('data/tmn')
    tmx_list = location.findall('data/tmx')

    for city in city_list:
        print(f'<{city.text}>')
        for tmEf, wf, tmn, tmx in zip(tmEf_list, wf_list, tmn_list, tmx_list):      
            print(tmEf.text, wf.text, '   ', tmn.text, '~', tmx.text, '도')
```
