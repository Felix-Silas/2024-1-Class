```python
import seaborn as sns
import numpy as np



titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
import numpy as np
import pandas as pd

# 주어진 구간 경계값
bin = [0, 7.9104, 14.4542, 31.0, 512.3292]

count, bin_dividers = np.histogram(titanic['fare'], bins=bin)

print(count, bin_dividers)
```

```python
# optin: include_lowest = True 동작 확인
bin_names = ['D', 'C', 'B', 'A']

titanic['fare_abcd'] = pd.cut(
    x=titanic['fare'],
    bins=bin_dividers,
    labels=bin_names
)

for idx, col in enumerate(titanic['fare']):
    if col==46:
        print(titanic.iloc[idx])

# NaN 값
```
