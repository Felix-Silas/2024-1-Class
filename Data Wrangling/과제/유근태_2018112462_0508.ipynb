```python
import seaborn as sns

df = sns.load_dataset("titanic")
df
```

```python
# 1. embark_town 열에서 누락 데이터가 있는지 평가하여 있으면 그 개수를 출력하시오.
print('누락 데이터 개수:',df['embark_town'].isnull().sum())
```

```python
# 누락 데이터 중 앞에서부터 발견한 첫 번째 누락 데이터에 대해 그 앞뒤 2개씩 총 5개의 승선도시를 출력하시오.
missing_index = df['embark_town'].index[df['embark_town'].isnull()].tolist()[0]

print("누락된 데이터 앞뒤의 승선도시:")
print(df.loc[missing_index - 2 : missing_index + 2, 'embark_town'])
```

```python
# 승선 도시 중 가장 많이 승선한 도시를 찾아서 출력하시오.
embark_counts = df['embark_town'].value_counts()

most_common_embark_town = embark_counts.idxmax()

print("가장 많이 승선한 도시:", most_common_embark_town)
```

```python
# 승선도시 최빈값으로 모든 누락 데이터를 대체하고, 앞의 5개 데이터를 다시 출력하시오.
df['embark_town'].fillna(most_common_embark_town, inplace = True)

print(df.loc[missing_index - 2 : missing_index + 2, 'embark_town'])
```
