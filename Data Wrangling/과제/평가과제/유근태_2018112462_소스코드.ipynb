# 1. XML파일 수집 및 설명

```python
import xml.etree.ElementTree as ET

tree = ET.parse('./iskc.d_0001.xml')

root = tree.getroot()
```

```python
root
```

# 2. XML파싱

```python
level1_list = root.findall('level1')

list1 = []

for level1 in level1_list:
    maintitle_list = level1.findall('front/biblioData/title/mainTitle')
    alternative_list = level1.findall('front/biblioData/title/alternative')
    for maintitle, alternative in zip(maintitle_list, alternative_list):
        continue
    level2_list = level1.findall('level2')
    level2_maintitle_list = level1.findall('level2/front/biblioData/title/mainTitle')
    for level2, level2_maintitle in zip(level2_list, level2_maintitle_list):
        att = level2.attrib
        attributes = level2_maintitle.attrib
        id = att.get('id')
        volume = attributes.get('volume')
        list1.append([maintitle.text, alternative.text, level2_maintitle.text, volume, id])


'''
이 Python 코드는 XML 파일을 파싱하여 특정 데이터를 추출하고 리스트에 저장하는 작업을 수행합니다. 먼저 루트 요소에서 'level1' 태그를 가진 모든 요소를 찾고, 추출한 데이터를 저장할 빈 리스트를 생성합니다. 각 'level1' 요소에 대해 'front/biblioData/title/mainTitle'과 'front/biblioData/title/alternative' 태그를 가진 요소들을 찾습니다. 그런 다음, 각 'level1' 요소 내에서 'level2' 요소와 그 하위의 'front/biblioData/title/mainTitle' 요소를 찾습니다. 두 리스트를 동시에 순회하면서 각 'level2' 요소의 속성과 'level2_maintitle' 요소의 속성을 추출합니다. 여기서 'id'와 'volume' 속성 값을 가져오고, 마지막으로 'maintitle', 'alternative', 'level2_maintitle', 'volume', 'id' 값을 리스트로 묶어 list1에 추가합니다.
'''
```

```python
import pandas as pd
df = pd.DataFrame(list1, columns=['maintitle', 'alternative', 'level2_maintitle', 'volume', 'id'])
df
```

# 3. 누락 데이터 평가 및 구현

```python
# 누락 데이터이면 True 반환, 유효한 데이터이면 False 반환
df.isnull()
```

```python
# 컬럼별 null 값의 개수 구하기
missing_values = df.isnull().sum(axis = 0)

print(missing_values)
```

```python
# 컬럼별 null 값 percent 구하기
missing_percentage = (missing_values / len(df)) * 100

print(missing_percentage)
```
