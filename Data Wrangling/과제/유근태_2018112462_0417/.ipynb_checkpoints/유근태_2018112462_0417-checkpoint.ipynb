```python
import pandas as pd

df = pd.read_csv('data-text.csv')

df
```

```python
# iloc 인덱서 & slicing
df100 = df.iloc[:100]

df100
```

```python
df100.to_csv('data-text100.csv')
```

```python
df2 = df100[['Indicator', 'Year', 'Country', 'Sex']][-20:]
df2.reset_index(inplace = True)
df2['Year'][1] = 2002

df_final = pd.DataFrame(df2)
df_final.reset_index(inplace = True)


df_final
```

```python
import sqlite3

#1. DB 연결 (파일이 없으면 만들고 있으면 연결)
con = sqlite3.connect("C:/Temp/sqlite3/pand.db")

# 2. DB 저장
df_final.to_sql('life', con, if_exists = 'append', index = False)

# 3. DB 연결 종료
con.close()
```
