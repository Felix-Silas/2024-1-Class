```python
from csv import reader

# 읽기
data_rdr = reader(open('mn_cut100.csv', 'rt'))
header_rdr = reader(open('mn_headers.csv', 'rt', encoding = 'utf8'))

# list comprehension
data_rows = [d for d in data_rdr]
header_rows = [h for h in header_rdr]

# 두 헤더의 길이 출력
print(len(data_rows[0]))
print(len(header_rows))
```

```python
# 데이터 헤더 행 (헤더 시작부분 눈으로 확인)
data_rows[0]
```

```python
# 헤더 시작 2행
header_rows[:2]
```

```python
# bad_rows에 데이터와 일치하지 않는 헤더 목록 만들기
bad_rows = []

for h in header_rows:
    if h[0] not in data_rows[0]:
        bad_rows.append(h)

len(bad_rows)

# 150개의 교집합, 60개의 불량 헤더목록 데이터, mismatch 데이터 9개
# 9개는 어떻게 할 것인가? --> 9개 처리하기 힘들면 버리거나, 중요한 정보면 어떻게 살릴 것인가를 판단(ex) header 파일에 추가)
# 여기선 2개 살리고, 7개 버린다. 웹스크래핑해서 찾아야겠지만, 그건 힘드니 다른 파일에 2개에 대한 설명이 들어 있게 했음
```

```python
# 헤더(설명) 파일에서 일치하지 않는 헤더 제거
for h in bad_rows:
    header_rows.remove(h)
len(header_rows)
```

```python
# 정리된 헤더 목록에 없는 데이터 헤더 목록 찾기
all_short_headers = [h[0] for h in header_rows]
for header in data_rows[0]:
    if header not in all_short_headers:
        print('mismatch', header)

# 대문자로 있는 2개의 label은 의미가 있으니 살리고, 소문자로 되어 있는 7개의 레이블은 버린다.
```

```python
# mn_headers.csv는 210개였다면, mn_headers_updated.csv는 대문자 2개의 레이블이 추가된 212개로 구성

data_rdr = reader(open('mn_cut100.csv', 'rt'))
header_rdr = reader(open('mn_headers_updated.csv', 'rt', encoding = 'utf8'))

# list comprehension
data_rows = [d for d in data_rdr]
# 데이터에 없는 헤더는 제거하고 헤더 리스트를 만든다.
header_rows = [h for h in header_rdr if h[0]in data_rows[0]]

# 두 헤더의 길이 출력
print(len(data_rows[0]))
print(len(header_rows))
```

```python
# mismatch 7개 데이터 index 추출
skip_index = []
all_short_headers = [h[0] for h in header_rows]
for header in data_rows[0]:
    if header not in all_short_headers:
        index = data_rows[0].index(header)
        skip_index.append(index)
skip_index
```

```python
# mismatch 삭제 159 --> 152
new_data = []
for row in data_rows:
    new_row = []
    for i, d in enumerate(row):
        if i not in skip_index:
            new_row.append(d)
    new_data.append(new_row)

len(new_data[0])
```

```python
# 데이터에ㅓ mistmatch 7개를 삭제한 data_headers 만들기
data_headers = []
for i, header in enumerate(data_rows[0]):
    if i not in skip_index:
        data_headers.append(header)
len(data_headers)
```

```python
# 어긋나는 위치 확인
# ('MMC1', 'MTA1')부터 어긋남
for i, header in enumerate(data_headers):
    if header not in all_short_headers[i]:
        print(header, all_short_headers[i])
```

```python
skip_index = []
final_header_rows = []
all_short_headers = [h[0] for h in header_rows]

for header in data_rows[0]:
    if header not in all_short_headers:
        index = data_rows[0].index(header)
        skip_index.append(index)
        # 어긋나는 부분 재배열
    else:
        for head in header_rows:
            if head[0] == header:
                final_header_rows.append(head)
                break
skip_index
```

```python
# 헤더 설명 list 만들기
header1 = []
for row in final_header_rows:
    header1.append(row[1])
header1
```

```python
# 데이터에 헤더설명 추가
new_data.insert(1, header1)
```

```python
# list --> dataframe csv 저장
import pandas as pd

df = pd.DataFrame(new_data)
df.to_csv('mn_clean02.csv', index = False)
```
