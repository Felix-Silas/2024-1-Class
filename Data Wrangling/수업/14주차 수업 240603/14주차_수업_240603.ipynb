```python
import pandas as pd

df = pd.read_csv('./auto-mpg.csv', header=None)
df.columns =  ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'name']
df
```

```python
mpg2kpl = 1.60934 / 3.78541

mpg2kpl
```

```python
df['kpl'] = df['mpg'] * mpg2kpl
df['kpl'] = df['kpl'].round(2)
df
```

```python
df.info()
```

```python
print(df['horsepower'].unique())
```

```python
import numpy as np
df['horsepower'].replace('?', np.nan, inplace=True)

df.info()
```

```python
df.dropna(subset=['horsepower'], inplace=True)
df.info()
```

```python
df['horsepower'] = df['horsepower'].astype('float')

print(df['horsepower'].dtypes)
```

```python
count, bin_dividers = np.histogram(df['horsepower'], bins=3)

print(count, bin_dividers)
```

```python
# optin: include_lowest = True 동작 확인
bin_names = ['저출력', '보통출력', '고출력']

df['hp_bin'] = pd.cut(
    x=df['horsepower'],
    bins=bin_dividers,
    labels=bin_names
)

for idx, col in enumerate(df['horsepower']):
    if col==46:
        print(df.iloc[idx])

# NaN 값
```

```python
bin_names = ['저출력', '보통출력', '고출력']
df['hp_bin'] = pd.cut(
    x=df['horsepower'],
    bins=bin_dividers,
    labels=bin_names,
    include_lowest=True # 여기를 빼고 돌리면
)
df[['horsepower', 'hp_bin']].head(15)
```

```python
def label_encoding(data):
    if data == '저출력':
        return 0
    elif data == '보통출력':
        return 1
    else:
        return 2


df['hp_le'] = df['hp_bin'].apply(label_encoding)
df[['horsepower', 'hp_bin', 'hp_le']].head(15)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['hp_le'] = le.fit_transform(df['hp_bin'])
df[['horsepower', 'hp_bin', 'hp_le']].head(15)
```
