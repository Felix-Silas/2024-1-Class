```python
from csv import reader

# 읽기
data_rdr = reader(open('mn_cut100.csv', 'rt'))
header_rdr = reader(open('mn_headers.csv', 'rt', encoding = 'utf8'))

data_rows = [d for d in data_rdr] # data_rows[0]이 header임
header_rows = [h for h in header_rdr]

# 데이터와 일치하지 않는 헤더 목록 만들기
bad_rows = []

for h in header_rows:
    if h[0] not in data_rows[0]:
        bad_rows.append(h)

# 헤더(설명) 파일에서 일치하지 않는 헤더 제거
for h in bad_rows:
    header_rows.remove(h)

# 정리된 헤더 목록에 없는 데이터 헤더 목록 찾기
all_short_headers = [h[0] for h in header_rows] # h[0]가 짧은 header 이름

for header in data_rows[0]: # data_rows[0]은 집합이고, 그 header들이 안에 있는지 체크
    if header not in all_short_headers:
        print('mismatch', header)

#print(data_rows[0])

print(len(header_rows))
```
