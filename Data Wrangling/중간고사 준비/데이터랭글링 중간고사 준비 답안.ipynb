```python
# 2-1
import csv

f = open('seoul.csv')
seoul = csv.reader(f)
header_seoul = next(seoul)
print(header_seoul)
```

```python
# 2-2
min_temp_seoul = 999

for row in seoul:
    row[0] = row[0].strip()
    if row[-2] == '':
        row[-2] = 999
    row[-2] = float(row[-2])
    if min_temp_seoul > row[-2]:
        min_temp_seoul = row[-2]
        min_date_seoul = row[0]

print(min_temp_seoul)
print(min_date_seoul)
```

# 3-2. friendship_pairs을 아래 포맷으로 출력
[(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]

```python
import csv

pair_list = []

with open('friendship_pairs.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        pair_list.append((int(row[0]), int(row[1])))

print(pair_list)
```

# 3-3. 이전 2개의 데이터를 활용하여 id를 key로 사용하고, 해당 id의 모든 친구 목록을 값으로 구성한 firendship 딕셔너리를 생성하시오.
{0: [1, 2], 1: [0, 2, 3], 2: [0, 1, 3], 3: [1, 2, 4], 4: [3, 5], 5: [4, 6, 7], 6: [5, 8], 7: [5, 8], 8: [6, 7, 9], 9: [8]}


```python

dict = {}

for a, b in pair_list:
    if a not in dict:
        dict[a] = []
    if b not in dict:
        dict[b] = []
    dict[a].append(b)
    dict[b].append(a)

print(dict)
    
```

```python
friendship= {}
# 각 사용자의 친구 목록을 구성
for user_id, _ in enumerate(users):
    friendship[user_id] = []

# pair_list를 사용하여 친구 관계를 설정
for pair in pair_list:
    user1, user2 = pair
    friendship[user1].append(user2)
    friendship[user2].append(user1)

# friendship 딕셔너리 출력
print(friendship)
```

# 4. mid-term-rss3.xml을 분석하여, 아래 도시명과 날씨 정보들을 xml에서 파싱하고, 아래 포맷으로 출력되도록 프로그램을 작성하시오.

```python
import xml.etree.ElementTree as ET

tree = ET.parse('mid-term-rss3.xml')

root = tree.getroot()
```

```python
root.tag
```

```python
wf_list = root.findall('channel/item/description/body/location/data/wf')

wf_list
```

```python
city_list = root.findall('channel/item/description/body/location/city')
tmEf_list = root.findall('channel/item/description/body/location/data/tmEf')
wf_list = root.findall('channel/item/description/body/location/data/wf')
tmn_list = root.findall('channel/item/description/body/location/data/tmn')
tmx_list = root.findall('channel/item/description/body/location/data/tmx')

for city in city_list:
    print('<',city.text,'>')
    for tmEf, wf, tmn, tmx in zip(tmEf_list, wf_list, tmn_list, tmx_list):
        print(tmEf.text, wf.text,'\t' ,tmn.text,'~',tmx.text, '도')
    
```

# 5-2. 20행 csv 파일을 읽어, 이 중 'Indicator', 'Country', 'Display Value' 필드만 추출하여 SQLite 데이터베이스에 저장하시오.

```python
import sqlite3

con = sqlite3.connect('C:/Temp/sqlite3/test.db')

cursor = con.cursor()

cursor.execute("CREATE TABLE practice(Indicator, Country, Display Value)")

with open('data-text20.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        cursor.executre("INSERT INTO practice VALUES(?, ?, ?)", row[0], row[5], row[7])

con.commit()

con.close()
```
