# 1번

```python
import xml.etree.ElementTree as ET


tree = ET.parse('hospital.xml') 

root = tree.getroot() 
```

```python
name_list = root.findall('Staff/Doctors/Doctor/DoctorName')
spe_list = root.findall('Staff/Doctors/Doctor/Speciality')
add_list = root.findall('Staff/Doctors/Doctor/DoctorContact/DoctorAddress')
rel_list = root.findall('Staff/Doctors/Doctor/Relantionship')
doc_list = root.findall('Staff/Doctors/Doctor')

doctor_list = []

for name, spe, add, rel, doc in zip(name_list, spe_list, add_list, rel_list, doc_list):
    doctor_list.append([name.text, spe.text ,add.text, rel.get('RelantionshipType'), doc.get('WorksIn')])
    #print(doc.get('WorksIn'))

doctor_list
```

```python
import json

list2 = []

with open('doctors.json') as f:
    ob =json.load(f)
    for o in ob:
        #print(o)
        #print(o.values())
        a, b = o.values()
        list2.append([a, b])

list2
```

```python
doctor_list2 = []

for doctor in doctor_list[:21]:
    for name, age in list2:
        if doctor[0] == name:
            # print(doctor)
            doctor.append(age)
            #print(doctor)
            doctor_list2.append(doctor)
        
doctor_list = doctor_list2

doctor_list
```
