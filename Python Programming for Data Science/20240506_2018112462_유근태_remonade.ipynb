# 레모네이드 판매예측
## deep learning 기초 1

```python
# 1. 라이브러리 불러오기
import tensorflow as tf
import pandas as pd
```

```python
# 2. 파일 불러오기 / 영어만인지, 한글 포함인지 조심(utf-8)
file_path = './lemonade1-1_E.csv'
data = pd.read_csv(file_path) # 변수명 주의
data.head()
```

```python
# 3. # 파일 데이터 분리 (독립 = indeopen, 종속 = deopen)
indeopen = data[['TEMP']] # 독립 변수
deopen = data[['SALES']] # 종속 변수
print(indeopen.shape, deopen.shape)
```

```python
# 4. 모델 만들기
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)

model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 5. 모델 학습 (FIT)
model.fit(indeopen, deopen, epochs=3000) # 3000번 학습
```

```python
# loss = 0.0116 ... 나쁘지 않음 => 0.0002 )
model.predict(indeopen)
```

```python
# 우리가 예측하고 싶은 값을 넣음 1
model.predict([[15]])
```

```python
# 우리가 예측하고 싶은 값을 넣음 2
model.predict([[30]])
```
