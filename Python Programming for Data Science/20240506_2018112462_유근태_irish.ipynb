# iris = 붓꽃

```python
# 1. 라이브러리 불러오기
import tensorflow as tf
import pandas as pd
```

```python
# 2. 파일 불러오기 / 영어만인지, 한글 포함인지 조심(utf-8)
file_path = './iris_E.csv'
data = pd.read_csv(file_path)
```

```python
data.head(10)
```

```python
data.tail()
```

```python
# 2-2. 원핫인코딩
incod = pd.get_dummies(data)
print(incod.columns)
```

```python
incod.head()
```

```python
# 3. 독립 변수 = indep, 종속 = depen
indeopen = incod[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
deopen = incod[['variety_setosa', 'variety_versicolor', 'variety_virginica']]
print(indeopen.shape, deopen.shape)
```

```python
# 4. 모델 만들기
X = tf.keras.layers.Input(shape=[4]) # 변수 4개니까 4 들어감
Y = tf.keras.layers.Dense(3, activation = 'softmax')(X) # activation 다양함

model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics = 'accuracy') # accracy 지우면 loss가 기본값
```

```python
# 5. 모델 학습 (FIT)
model.fit(indeopen, deopen, epochs=1000) # 1000번 학습
```

```python
# 우리가 예측하고 싶은 값을 넣음 1
model.predict(indeopen[0:5])
```

```python
# 우리가 예측하고 싶은 값을 넣음 2
model.predict(indeopen[-5:])
```
