# -4) 도미와 빙어 데이터를 하나의 데이터로 합침

```python
# 데이터 합치기

length = bream_length + smelt_length
weight = bream_weight + smelt_weight

# 리스트 형태로 만들기
fish_data = [[l, w] for l, w in zip(length, weight)]

print(fish_data)
```

# -5) 도미: [1], 빙어: [0] => 도미와 빙어를 숫자 1, 0으로 표현

```python
# 도미와 빙어를 숫자 1, 0으로 표현
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
# for문으로 fish_target 저장
fish_target = []

for i in range(len(bream_length)):
  fish_target.append(1)

for i in range(len(smelt_length)):
  fish_target.append(0)

print(fish_target)
print(len(fish_target))
```

# -6) KNN (K-Nearleast Neighbors)
## 단점: 데이터가 500~1000개가 넘는, 데이터가 많은 경우는 사용할 수 없다!

```python
'''import sklearn

model = sklearn.neighbors.KNeighborsClassifier()
model.fit(fish_data, fish_target)'''
```

```python
# 사이킷런 KNN 라이브러리
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()

# 모델 학습
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target)
```

# -7) 30 length, 600 weight는? 13 length, 15  weight는?

```python
kn.predict([[30, 600]])

# 1: 도미로 예측
```

```python
kn.predict([[13, 15]])

# 0: 빙어로 예측
```
